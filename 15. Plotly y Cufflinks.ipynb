PLOTLY Y CUFFLINKS: VISUALIZACIONES INTERACTIVAS

```python
import pandas as pd
import numpy as np
```

```python
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
#Configuramos algunos parámetros
init_notebook_mode(connected=True)
cf.go_offline()
%matplotlib inline
```

```python
#Creamos un dataframe
dataframe= pd.DataFrame(np.random.randn(100,4), columns=['a', 'b', 'c', 'd'])
dataframe               
```

```python
#Empezamos creando el primer gráfico
dataframe.plot()
```

```python
#Con iplot creamos un gráfico interactivo que nos permite movernos por las distintas columnas.
#Podemos coger un trozo del gráfico y ampliarlo si lo seleccionamos. Al hacer doble click volvemos al gráfico original
#Podemos activar/desactivar columnas pulsando sobre ellas en la leyenda
#Nos permite guardar una immagen con lo que aparece ahora mismo en el gráfico con el boton de camara
dataframe.iplot()
```

```python
#Tb podemos hacr otro tipo de gráficos con iplot
dataframe.iplot(kind='scatter', x='a', y='b', mode='markers')
```

```python
dataframe.iplot(kind='bar')
```

```python
#Podemos hacer uno, por ejemplo, con la suma de los valores de cada columna:
dataframe.sum().iplot(kind='bar')
```

```python
dataframe.iplot(kind='box')
```

```python
Histograma de la columna a
dataframe['a'].iplot(kind='hist', bins=30)
```

```python
dataframe.iplot(kind='hist', bins=30)
```

```python
#Diagramas spread spbre 2 columnas:
dataframe[['a','b']].iplot(kind='spread')
```

```python
#Diagrama de tipo bubble
dataframe.iplot(kind='bubble', x='a', y='b', size='c')
```

```python
#Gráfico en 3d
#creamos un nuevo dataframe
dataframe2= pd.DataFrame({'a':[1,2,3,4], 'b':[30,40,20,10], 'c':[12,16,18,15]})
dataframe2
```

```python
dataframe2.iplot(kind='surface')
```
