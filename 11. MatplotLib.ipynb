```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creamos un array con nº entre el 2 y el 9 que contenga 10 elementos
x = np.linspace(2,9,10)
x
```

```python
# La serie y será basada en x al cuadrado
y = x**2
y
```

```python
#Hacemos un gráfico con los dos arrays
plt.plot(x,y)
plt.show()
```

```python
#Podemos cambiar el color de la gráfica
plt.plot(x,y,'red')
```

```python
#Podemos poner el color que queramos
plt.plot(x,y,'lightblue')
```

```python
#Tb podemos añadirle un  título:
plt.plot(x,y,'red')
plt.title('Este es mi gráfico')
```

```python
#Podemos ponerle valor al eje x y al eje y
plt.plot(x,y,'red')
plt.title('Este es mi gráfico')
plt.xlabel("Eje x")
plt.ylabel('Eje y')
```

```python
#Podemos añadir varios gráficos en el mismmo espacio. (1,2,1) el primer 1 significa que va a tener 1 fila, el 2 que va a tener 2 columnas, 
# y el ultimo uno significa la posición del gráfico
plt.subplot(1,2,1)
plt.plot(x,y,'green')

plt.subplot(1,2,2)
plt.plot(y,x,'red')
```

```python
#Si queremos poner mejor 2 filas y 1 columna:
plt.subplot(2,1,1)
plt.plot(x,y,'green')

plt.subplot(2,1,2)
plt.plot(y,x,'red')
```

```python
#Podemos hacerlo tb mediante la programación orientada a objetas:
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.8,0.8])
grafico.plot(x,y,'green')
```

```python
#Podemos poner tb 2 gráficos:
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,'r')

grafico2 = figura.add_axes([0.2,0.5,0.3,0.3])
grafico2.plot(y,x,'b')
```

```python
#Multigráficos con distinto nº de filas y de columnas
# Vamos a construir un gráfico de 4 subgráficos organizados en forma de 2 filas y 2 columnas
figura, gráficos = plt.subplots(nrows=2, ncols=2)
```

```python
#Vamos a rellenar cada gráfico:
figura, gráficos = plt.subplots(nrows=2, ncols=2)
gráficos[0][0].plot(x,y,'blue')
gráficos[0][1].plot(x,y,'red')
gráficos[1][0].plot(x,y,'green')
gráficos[1][1].plot(x,y,'y')
#Podemos añadirles título:
gráficos[0][0].set_title('Azul')
gráficos[0][1].set_title('Rojo')
gráficos[1][0].set_title('Verde')
gráficos[1][1].set_title('Amarillo')
```

```python
# Definir el tamaño de la figura/gráfico:
figura = plt.figure(figsize=(8,4))
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r')
```

```python
#Crear 2 gráficos juntos:
#1º creamos la figura
figura = plt.figure()
graficos = figura.add_axes([0.1,0.1,0.9,0.9])
graficos.plot(x,x**2,label='Al cuadrado')
graficos.plot(x,x**3, label='Al cubo')
#Añadimos una leyenda para saber cuál es cada gráfica:
graficos.legend(loc=0)
```

```python
#Color del gráfico
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color='red')
```

```python
#Tipo de línea del gráfico:
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color='red', linewidth=5)
```

```python
#Para hacer la línea más transparente:
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color='red', linewidth=5, alpha=0.5)
```

```python
#Para cambiar el tipo de línea:
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color='red', linewidth=10, alpha=0.2, linestyle=':')
```
