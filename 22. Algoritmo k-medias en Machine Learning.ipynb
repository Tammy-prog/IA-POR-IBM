ALGORITMO K-MEDIAS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creaamos un dataframe con make_bolbs (n_samples: nº de ejemplos, n_features: nº de características, centers: nº de grupos)
from sklearn.datasets import make_blobs
datos = make_blobs(n_samples=200, n_features=2, centers=4)
datos
```

```python
# En datos[0] vemos los datos y en datos[1] los grupos
datos[0]
```

```python
datos[1]
```

```python
#Hacemos un gráfico para verlo más claro
plt.scatter(datos[0][:,0], datos[0][:,1])
```

```python
# Vamos a crear un algoritmo de k-medias para, dados los datos (datos[0]) poder agruparlos en el nº de grupos que nosotrso queramos
from sklearn.cluster import KMeans
modelo = KMeans(n_clusters=4)
modelo.fit(datos[0])
```

```python
modelo.cluster_centers_
```

```python
modelo.labels_
```

```python
datos[1]
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,4))

ax1 = plt.scatter(datos[0][:,0], datos[0][:,1], c= modelo.labels_)
#ax1.set_title('Algoritmo de k-medias')

ax2 = plt.scatter(datos[0][:,0], datos[0][:,1], c= datos[1])
#ax1.set_title('Datos originales')
```
