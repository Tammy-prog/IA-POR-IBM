FACETGRID

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Cargamos de nuevo el dataset propinas de seaborn
propinas= sns.load_dataset('tips')
propinas.head()
```

```python
# Para saber cuantos elementos distintos hay por ejemplo de time dentro de nuestro dataset:
propinas['time'].unique()
```

```python
propinas['sex'].unique()
```

```python
propinas['day'].unique()
```

```python
graficos2 = sns.FacetGrid(data=propinas, col='time', row='smoker')
graficos2.map(sns.distplot,'total_bill')
```

```python
# Si queremos usar otro tipo de gráfico en vez de distplot:
graficos2 = sns.FacetGrid(data=propinas, col='time', row='smoker')
graficos2.map(plt.scatter,'total_bill', 'tip')
```

GRÁFICOS DE REGRESION (REGRESSION PLOTS)

```python
# Vamos a volver a usar propinas
propinas.head()
```

```python
#Gráfico de regresión 
sns.lmplot(x='total_bill', y='tip', data=propinas)
```

```python
#Podemos añadirle una vble nueva:
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='sex')
```

```python
#Podemos cambiar los marcadores:
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='sex', markers=['o','x'])
```

```python
#Podemos ampliar el gráfico para que se vea mejor (size =80):
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='sex', markers=['v','x'], scatter_kws={'s':60})
```

```python
# Ahora vamos a ver como, parrt8iendo del original (lmplot) como podemos añadirle columnas:
sns.lmplot(x='total_bill', y='tip', data=propinas, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, col='day')
```

```python
#Tb podemos hacerlo con las filas:
sns.lmplot(x='total_bill', y='tip', data=propinas, col='day', row='sex')
```

```python
# En vez de row tb podríamos poner hue y dentro del mismo gráfico nos separara por el parámetro que le pongamos:
sns.lmplot(x='total_bill', y='tip', data=propinas, col='day', hue='sex')
```

```python
# Volvemos al original y podemos cambiar tb el aspecto del gráfico (aspect es la relación entre ancho y alto, height sustituye a size):
sns.lmplot(x='total_bill', y='tip', data=propinas, aspect=0.8, height=10)
```

ESTILOS Y COLORES EN SEABORN

```python
propinas.head()
```

```python
sns.countplot(x='sex', data=propinas)
```

```python
#Si queremos cambiar el estilo, tenemos que añadir delante:
sns.set_style('darkgrid')
sns.countplot(x='sex', data=propinas)
```

```python
#Para quitar el recuadro de alrededor:
sns.set_style('white')
sns.countplot(x='sex', data=propinas)
sns.despine()
```

```python
#Podemos borrar tb la barra izqda y la del fondo:
sns.set_style('white')
sns.countplot(x='sex', data=propinas)
sns.despine(left=True, bottom=True)
```

```python
#Cambiamos el tamaño:
plt.figure(figsize=(12,4))
sns.set_style('white')
sns.countplot(x='sex', data=propinas)
sns.despine(left=True, bottom=True)
```

```python
#Podemos modificar el tamaño de las letras:
sns.set_context('notebook', font_scale=1.5)
plt.figure(figsize=(12,4))
sns.set_style('white')
sns.countplot(x='sex', data=propinas)
sns.despine(left=True, bottom=True)
```

```python
#Para cambiar los colores (buscamos las paletas que hay en la página de seaborn):
sns.countplot(x='sex', data=propinas, palette='Set2')
```

```python
sns.countplot(x='sex', data=propinas, palette='husl')
```

```python
#Podemos elegir nosotros los colores que queremos:
colores = ['lightgreen','blue']
sns.countplot(x='sex', data=propinas, palette=colores)
```

```python
#Tb podemos poner los colorees con el código de color:
colores = ['#24CA9E','#090847']
sns.countplot(x='sex', data=propinas, palette=colores)
```
