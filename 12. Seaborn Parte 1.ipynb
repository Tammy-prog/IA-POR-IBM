SEABORN

```python
import seaborn as sns
```

```python
#Vamos a cargar un dataset que viene por defecto en la librería para poder hacer pruebas
propinas = sns.load_dataset('tips')
```

```python
propinas.head(10)
```

```python
#Hacemos unn histograma sns sobre la primera columna 'total_bill'
sns.displot(propinas['total_bill'])
```

```python
# Si queremos configurar en nº de barras:
sns.displot(propinas['total_bill'], bins=30)
```

```python
sns.displot(propinas['total_bill'], bins=50)
```

```python
sns.displot(propinas['tip'], bins=30)
```

```python
#Para poder comparar dos variables en un gráfico:
sns.jointplot(x='total_bill', y='tip', data= propinas)
```

```python
#Podemos cambiar el tipo de gráfico, y poner, x ejempo, uno hexadecimal
sns.jointplot(x='total_bill', y='tip', data= propinas, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data= propinas, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data= propinas, kind='kde')
```

```python
#Para poder comparar todas las columnas que sean numéricas:
sns.pairplot(propinas)
```

```python
# A estos gráficos podemos añadirle tb columnas que no sean numéricoas, sino tipo category:
sns.pairplot(propinas, hue='sex')
```

```python
sns.pairplot(propinas, hue='size')
```

```python
# Gráfico rugplot: muestra mediante una raya cada dato del dataset
sns.rugplot(propinas['total_bill'])
```

```python
sns.rugplot(propinas['size'])
```

```python
sns.rugplot(propinas['tip'])
```

GRÁFICOS PARA COLUMNAS TIPO CATEGORY

```python
import seaborn as sns
import numpy as np
```

```python
propinas.head(10)
```

```python
sns.barplot(x='sex', y='total_bill', data=propinas)
```

```python
#Para contar el nº de personas masculinas o femeninas:
sns.countplot(x='sex', data= propinas)
```

```python
sns.countplot(x='smoker', data= propinas)
```

```python
#Gráfico para comparar ujna columna category con una columna no numérica:
sns.boxplot(x='day', y='total_bill', data=propinas)
```

```python
#Podemos añadirle otra vble más, por ejemplo si es fumador:
sns.boxplot(x='day', y='total_bill', data=propinas, hue='smoker')
```

```python
#Podemos hacerlo tb de tipo violinplot
sns.violinplot(x='day', y='total_bill', data=propinas)
```

```python
sns.boxplot(x='day', y='total_bill', data=propinas, hue='smoker')
```

```python
#Gráfico que compara todos los puntos del dataset:
sns.stripplot(x='day', y='total_bill', data=propinas)
```

```python
sns.stripplot(x='day', y='total_bill', data=propinas, hue ='sex')
```

```python
#Otro diagrama de este estilo:
sns.swarmplot(x='day', y='total_bill', data=propinas, hue ='sex')
```

```python
#Para unir dos gráficos simplemente los ejecutamos a la vez:
sns.violinplot(x='day', y='total_bill', data=propinas)
sns.swarmplot(x='day', y='total_bill', data=propinas)
```
