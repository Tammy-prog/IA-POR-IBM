```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Vamos a usar el documento 'vino.csv' que hemos guardado de nuestra carpeta del proyecto (C:\Users\34649\Documents\Machine Learning Data Science)
vinos = pd.read_csv('vino.csv')
vinos.head()
```

```python
#Vamos a ver losn distintos valores que posee la columna 'Wine Types'
vinos['Wine Type'].unique()
```

```python
#Podemos ver el nº de vinos de cada tipo:
vinos['Wine Type'].value_counts()
```

```python
# Separamos las variables de la vble que queremos predecir(Wine Type)
x= vinos.drop('Wine Type', axis=1)
y= vinos['Wine Type']
```

```python
# Dividimos el dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

```python
# Creamos y entrenamos el modelo
from sklearn.tree import DecisionTreeClassifier
arbol = DecisionTreeClassifier()
arbol.fit(x_train, y_train)
```

```python
# Creamos nuestras propias predicciones
predicciones = arbol.predict(x_test)
```

```python
predicciones
```

```python
#Estos son los valores reales, lo q vamos a comprarar con las predicciones
y_test
```

```python
# Evaluamos el modelo:
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predicciones))
```

```python
print(confusion_matrix(y_test, predicciones))
```
