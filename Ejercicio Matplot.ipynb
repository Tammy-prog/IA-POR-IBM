2. Crear una vble y que contenga los valores de la funcion x al cuadrado +5

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,20,10)
```

```python
y = (x**2)+5
```

```python
x
```

```python
y
```

```python
figura = plt.figure()
grafico= figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'green', linestyle=':', linewidth=5) 
```
