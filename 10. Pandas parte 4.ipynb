GRABAR DATAFRAME DENTRO DE UNA TABLA DE SQL DE BBDD

conda install sqlalchemy /  pip install sqlalchemy

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Creamos un dataframe a partir de un diccionario
diccionario = {'A': [1,2,3], 'B':[4,5,6]}
dataframe = pd.DataFrame(diccionario)
```

```python
dataframe
```

```python
#Ahora vamos a grabar este datafranme dentro de una bbdd
# 1º creamos en memoria un motor de bbdd para poder grabar el dataframe dentro de una tabla
engine = create_engine('sqlite:///:memory:')
```

```python
#Guardamos la tabla. El index = False significa que el índice (0,1,2) no lo grabe
dataframe.to_sql('tabla_ejercicio', engine, index=False)
```

```python
#Para comprobar que está bien (con = conexion):
datos_leidos_bd = pd.read_sql('tabla', con=engine)
datos_leidos_bd
```

GRÁFICOS CON PANDAS

```python
import numpy as np
import pandas as pd
```

```python
# Creamos un dataframe a través de la función de números aleatorios enteros de numpy
dataframe2 = pd.DataFrame(np.random.randint(200, size=(50,4)), columns=['a', 'b', 'c', 'd'])
```

```python
dataframe2
```

```python
#Creamos uin histograma con los valores de la columna 'a':
dataframe2['a'].hist()
```

```python
#Con la columna 'b'
dataframe2['b'].hist()
```

```python
#Podemos camb iar el nº de columnas que tiene nuestro histograma con bins
dataframe2['a'].hist(bins=30)
```

```python
#Podemos crearlo tb con otro método:
dataframe2['a'].plot.hist()
```

```python
#Gráfico de tipo área:
dataframe2.plot.area()
```

```python
#Poemos hacer que los colores sean más transparentes:
dataframe2.plot.area(alpha=0.3)
```

```python
#Podemos hacer tb un gráfico de barras
dataframe2.plot.bar()
```

```python
#Podemos hacer tb este diagrama de barras pero con los vaklores acumulados, es decir, que los valores se acumulen unos encima de otros
dataframe2.plot.bar(stacked=True)
```

```python
#Diagrama tipo scatter. Le pasamos las columnas que queramos para el gráfico:
dataframe2.plot.scatter(x='a', y='b')
```

```python
#Podemos añadirle más columnas. Segun la escala de grises es uno u otro
dataframe2.plot.scatter(x='a', y='b', c='c')
```

```python
#Podemos ponerle una paleta de color diferente con cmpap:
dataframe2.plot.scatter(x='a', y='b', c='c', cmap='coolwarm')
```

```python
#Podemos hacer gráfico de caja:
dataframe2.plot.box()
```

```python
#Gráfico hexadecimal:
dataframe2.plot.hexbin(x='a', y='b', gridsize=12)
```

```python
#Gráfico kde:
dataframe2.plot.kde()
```

```python
#Podemos hacer este mismo gráfico con el método density:
dataframe2.plot.density()
```
