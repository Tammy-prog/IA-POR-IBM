MERGE EN DATAFRAME

```python
import numpy as np
import pandas as pd
```

```python
#Creamos un primer dataframe a partir de un diccionsrio
diccionario = {'A':[1,2,3], 'B':[4,5,6], 'clave':['c1','c2', 'c3'] }
dataframe1 = pd.DataFrame(diccionario)
```

```python
dataframe1
```

```python
#Creamos un segundo dataframe
diccionario2 = {'A':[11,12,13], 'B':[14,15,16], 'clave':['c1','c2', 'c3'] }
dataframe2 = pd.DataFrame(diccionario2)
```

```python
dataframe2
```

```python
pd.merge(dataframe1, dataframe2, on= 'clave')
```

JOIN EN DATAFRAMES

```python
diccionario3 = {'A':[1,2,3], 'B':[4,5,6] }
dataframe3 = pd.DataFrame(diccionario3, index =['i1', 'i2', 'i3'])
dataframe3
```

```python
diccionario4 = {'C':[11,12,13], 'D':[14,15,16] }
dataframe4 = pd.DataFrame(diccionario4, index =['i1', 'i2', 'i3'])
dataframe4
```

```python
# Como los dataframes tienes los mismos índices podemos hacer un join
dataframe3.join(dataframe4)
```

```python
#Si tenemos un dataframe con más datos que otro, nos pone el resto de valores a nulo:
diccionario5 = {'A':[1,2,3,4], 'B':[4,5,6,7] }
dataframe5 = pd.DataFrame(diccionario5, index =['i1', 'i2', 'i3','i4'])
dataframe5
```

```python
dataframe5.join(dataframe4)
```

OTRAS OPERACIONES

```python
diccionario6 = {'A':[11,12,10,12], 'B':[14,18,16,17] }
dataframe6 = pd.DataFrame(diccionario6, index = ['i1', 'i2', 'i3', 'i4'])
dataframe6
```

```python
#Podemos obtener los valores únicos de 1 columna
dataframe6['A'].unique()
```

```python
#Podemos saber el nº de valores únicos que hay
dataframe6['A'].nunique()
```

```python
#Para saber elo nº de veces que aparece cada valor:
dataframe6['A'].value_counts()
```

```python
#Definimos una función que dado un nº devuelva ese nº multiplicado por 2
def multiplicar(x):
    return x*2
```

```python
#Si aplicamos la función multiplicar a nuestro dataframe nos multiplica cada valor por 2
dataframe6['A'].apply(multiplicar)
```

```python
#Podemos obtener el mismo resultado con la función lambda
dataframe6['A'].apply(lambda x:x*2)
```

```python
dataframe6
```

```python
#Queremos borrar la columna B
dataframe6.drop('B', axis=1)
```

```python
#Si queremos borrar 1 fila:
dataframe6.drop('i1')
```

```python
# Para saber el nº de columnas del dataframe
dataframe6.columns
```

```python
#Para saber el nº de filas
dataframe6.index
```

```python
#Ordenamos por la columna B
dataframe6.sort_values('B')
```
